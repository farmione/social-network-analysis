# Read Dataset

```python
# Install NetworkX and python-louvain
!pip install networkx python-louvain

import networkx as nx
import matplotlib.pyplot as plt

# Specify the path to the dataset file in your Google Drive
dataset_file = "/content/drive/My Drive/social_network_analysis/facebook_combined.txt.gz"
```

# Network metrices

```python
# Load the dataset into a NetworkX graph
G = nx.read_edgelist(dataset_file, create_using=nx.Graph(), nodetype=int)

# Number of nodes and edges
num_nodes = len(G.nodes())
num_edges = len(G.edges())

# Check basic graph statistics
print("Number of nodes:", len(G.nodes()))
print("Number of edges:", len(G.edges()))

# Average degree
avg_degree = sum(dict(G.degree()).values()) / num_nodes
print("Average degree:", avg_degree)

# Density
density = nx.density(G)
print("Density:", density)

# Degree distribution histogram
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
plt.hist(degree_sequence, bins=20, alpha=0.75)
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.title("Degree Distribution")
plt.show()
```

# Degree Centrality

```python
# Degree Centrality
degree_centrality = nx.degree_centrality(G)
print("Degree Centrality:", degree_centrality)

# Extract the degrees as a list
degrees = [degree_centrality[node] for node in G.nodes()]

# Node with the highest degree centrality
max_degree_centrality_node = max(degree_centrality, key=degree_centrality.get)
print("Node with the highest degree centrality:", max_degree_centrality_node)

# Visualize degree centrality using node size
plt.figure(figsize=(10, 8))
nx.draw(
    G,
    pos=nx.spring_layout(G, seed=42),
    node_size=[d * 1000 for d in degrees],
    node_color="skyblue",
    edge_color="gray",
    with_labels=False,
    alpha=0.5,
)
plt.title("Degree Centrality Visualization")
plt.show()
```

# Betweenness Centrality

```python
# Betweenness Centrality
betweenness_centrality = nx.betweenness_centrality(G)
print("Betweenness Centrality:", betweenness_centrality)

# Extract the betweenness centralities as a list
betweenness_centralities = [betweenness_centrality[node] for node in G.nodes()]

# Node with the highest betweenness centrality
max_betweenness_centrality_node = max(betweenness_centrality, key=betweenness_centrality.get)
print("Node with the highest betweenness centrality:", max_betweenness_centrality_node)

# Visualize betweenness centrality using node color
plt.figure(figsize=(10, 8))
nx.draw(
    G,
    pos=nx.spring_layout(G, seed=42),
    node_size=10,
    node_color=betweenness_centralities,
    cmap=plt.cm.viridis,
    edge_color="gray",
    with_labels=False,
    alpha=0.5,
)
plt.title("Betweenness Centrality Visualization")
plt.colorbar(label="Betweenness Centrality")
plt.show()
```

# Closeness Centrality

```python
# Closeness Centrality
closeness_centrality = nx.closeness_centrality(G)
print("Closeness Centrality:", closeness_centrality)

# Extract the closeness centralities as a list
closeness_centralities = [closeness_centrality[node] for node in G.nodes()]

# Node with the highest closeness centrality
max_closeness_centrality_node = max(closeness_centrality, key=closeness_centrality.get)
print("Node with the highest closeness centrality:", max_closeness_centrality_node)

# Visualize closeness centrality using node color
plt.figure(figsize=(10, 8))
nx.draw(G,pos=nx.spring_layout(G, seed=42),node_size=10,node_color=closeness_centralities,cmap=plt.cm.plasma,edge_color="gray",with_labels=False,alpha=0.5,)
plt.title("Closeness Centrality Visualization")
plt.colorbar(label="Closeness Centrality")
plt.show()
```

# Clustering coefficient

```python
# Clustering Coefficient
clustering_coefficient = nx.average_clustering(G)
print("Average Clustering Coefficient:", clustering_coefficient)
```

# Community Detect by (Louvin Algorithm)

```python
from community import community_louvain

# Detect communities using Louvain algorithm
partition = community_louvain.best_partition(G)
print("Communities:", partition)
```

# Shortest Path Analysis

```python
# Shortest path between two nodes
source_node = 1
target_node = 100
shortest_path = nx.shortest_path(G, source_node, target_node)
print(f"Shortest Path from Node {source_node} to Node {target_node}: {shortest_path}")

# Average shortest path length
avg_shortest_path_length = nx.average_shortest_path_length(G)
print("Average Shortest Path Length:", avg_shortest_path_length)
```

# Network Robustness by random node removal

```python
# Network robustness by random node removal
import random
num_nodes_to_remove = int(0.1 * num_nodes)  # Remove 10% of nodes
G_copy = G.copy()
random_nodes_to_remove = random.sample(list(G_copy.nodes()), num_nodes_to_remove)

G_copy.remove_nodes_from(random_nodes_to_remove)

# Calculate size of the largest connected component
largest_component_size = len(max(nx.connected_components(G_copy), key=len))
print("Size of Largest Connected Component after Random Node Removal:", largest_component_size)
```

# Connectivity Component

```python
# Connectivity components
components = list(nx.connected_components(G))
num_components = len(components)
print("Number of Connectivity Components:", num_components)
```

# Core-periphery analysis

```python
# Core-periphery analysis
core_periphery = nx.core.core_number(G)
print("Core Number:", core_periphery)
```
